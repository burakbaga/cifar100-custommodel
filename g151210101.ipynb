```python
import numpy as np 
import keras
import matplotlib.pyplot as plt

from custom_dnn import build_model
from custom_dnn import data_fit
from plot_graph import plot
```

```python
width,height,channel = (32,32,3)
num_of_classes = 6 
batch_size = 32
input_shape = (width,height,channel)
epoch = 20
train_path = "Dataset/train"
test_path = "Dataset/test"
```

### 1.BÖLÜM Sade Model (Dropout ve Augmentation Yok)

```python
model = build_model(False,num_of_classes,input_shape)
```

```python
history = data_fit(False,train_path,test_path,width,height,batch_size,epoch,model)
```

```python

acc = history.history["acc"]
val_accuracy = history.history["val_acc"]
loss = history.history["loss"]
val_loss = history.history["val_loss"]

plot(acc,val_accuracy,loss,val_loss,model_name="1") # kendi yazdığım kütüphaneden draw isimli fonksiyonu çağırıyorum.
```

### 2.KISIM (Dropout Kullanılan Model)

```python
model = build_model(True,num_of_classes,input_shape)
```

```python
history = data_fit(False,train_path,test_path,width,height,batch_size,epoch,model)
```

```python

acc = history.history["acc"]
val_accuracy = history.history["val_acc"]
loss = history.history["loss"]
val_loss = history.history["val_loss"]

plot(acc,val_accuracy,loss,val_loss,model_name="2")
```

### 3.Kısım Augmentation Kullanılan Model

```python
model = build_model(False,num_of_classes,input_shape)
```

```python
history = data_fit(True,train_path,test_path,width,height,batch_size,epoch,model)
```

```python

train_accuracy = history.history["acc"]
validation_accuracy = history.history["val_acc"]
train_loss = history.history["loss"]
validation_loss = history.history["val_loss"]

plot(train_accuracy,validation_accuracy,train_loss,validation_loss,model_name="3")
```

```python
from keras.datasets import cifar100

(trainX, trainy), (testX, testy) = cifar100.load_data()
```

```python
liste = []
classes = [19,20,33,35,86,91]
```

```python
for i in classes:
    liste.extend(list(np.where(trainy==i)[0][:10]))
```

```python
print('Train: X=%s, y=%s' % (trainX.shape, trainy.shape))
print('Test: X=%s, y=%s' % (testX.shape, testy.shape))
# plot first few images
k=0
plt.figure(figsize=(15,15))
for i in liste:
  
    plt.subplot(10,10,k+1)
    plt.imshow(trainX[i])
    k+=1
# show the figure
plt.savefig("classes.png")
plt.show()
```
